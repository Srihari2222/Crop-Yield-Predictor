```python
def change_case(i):
    i = i.replace(" ", "")
    i = i.lower()
    return i
```

```python
def remove_brackets(i):
    if '(' not in i:return i
    temp=''
    for j in i:
        if j=='(':return temp
        temp+=j
```

# Merging the fertilizer.csv data into main dataset

```python
crop.insert(3,'N',[0 for i in range(len(crop))])
```

```python
crop.insert(4,'P',[0 for i in range(len(crop))])
```

```python
crop.insert(5,'K',[0 for i in range(len(crop))])
```

```python
crop.insert(6,'pH',[0 for i in range(len(crop))])
```

```python
crop.head()
```

```python
fert.head()
```

```python
for idx,row in crop.iterrows():
    c=row['Crop']
    random_row=fert[fert['Crop']==c].sample()
    temp_values=random_row.values[0]
    crop.loc[idx,['N','P','K','pH']]=temp_values[1:5]
    
```
