# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("FinalDataSet_with_ph.csv")
```

```python
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df.head()
```

```python
df.corr(numeric_only = True)
```

```python
plt.scatter(df['Area_in_hectares'],df['Production_in_tons'])
plt.xlabel('Area_in_hectares')
plt.ylabel('Production_in_tons')
```

```python
plt.scatter(df['State_Name'],df['Production_in_tons'])
plt.xticks(rotation=90)
plt.xlabel('State_Name')
plt.ylabel('Production_in_tons')
```

```python
sns.regplot(data=df,x='Area_in_hectares',y='Production_in_tons')
```
